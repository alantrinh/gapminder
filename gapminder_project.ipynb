```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
path='./data'

df_life_expectancy = pd.read_excel(f'{path}/gapminder_lifeexpectancy.xlsx', index_col=0)
df_life_expectancy.index.name = 'country'
df_population = pd.read_excel(f'{path}/gapminder_population.xlsx', index_col=0)
df_population.index.name = 'country'
df_fertility = pd.read_csv(f'{path}/gapminder_total_fertility.csv', index_col=0)
df_fertility.index.name = 'country'
df_continents = pd.read_csv(f'{path}/continents.csv', sep=';', index_col=0)
```

```python
df_life_expectancy.columns = df_life_expectancy.columns.astype(int).tolist()
```

```python
df_population.columns = df_population.columns.astype(int).tolist()
```

```python
df_fertility.columns = df_fertility.columns.astype(int).tolist()
```

```python
# draw Histogram
df_life_expectancy[1900].dropna().hist(bins=20, facecolor='#6d0000', alpha=0.8, histtype='step');
plt.title('Life expectancy in 1900')
plt.xlabel('Life expectancy')
plt.ylabel('Number of countries')
plt.axis([15, 55, 0, 40])
```

```python
fig, ax = plt.subplots(figsize = (10,6))
ax.hist(df_life_expectancy[1900].dropna(), bins=10);

ax.set_xlabel('Life expectancy (years)', fontsize=14)
ax.set_ylabel('Number of countries', fontsize=14)
ax.set_title('Life expectancy in 1900')
```

```python
fig, ax = plt.subplots(figsize = (12,6))
ax.hist(df_population[1900].dropna(), bins=20);

ax.set_xlabel('Population', fontsize=14)
ax.set_ylabel('Number of countries', fontsize=14)
ax.set_title('Country populations in 1900')
```

```python
fig, ax = plt.subplots(figsize = (12,6))
ax.hist(df_fertility[1900].dropna(), bins=20);

ax.set_xlabel('Total Fertility', fontsize=14)
ax.set_ylabel('Number of countries', fontsize=14)
ax.set_title('Fertility in 1900')
```

```python
df_fertility_long = df_fertility.reset_index().melt(id_vars=['country'], var_name='year', value_name='fertility rate')
df_fertility_long
```

```python
df_fertility_long.pivot_table(columns='year', index='country', dropna=False)
```

```python
df_life_expectancy_long = df_life_expectancy.reset_index().melt(id_vars='country', var_name='year', value_name='life expectancy')
df_life_expectancy_long
```

```python
df_population_long = df_population.reset_index().melt(id_vars='country', var_name='year', value_name='population')
df_population_long['population'] = df_population_long['population'] / 1000000
df_population_long
```

```python
df_continents_long = df_continents.reset_index()
df_continents_long
```

```python
pd.concat([df_fertility_long, df_life_expectancy_long]) # just concatenating the tables
```

```python
df_fertility_long.join(df_life_expectancy_long, lsuffix='_fert', rsuffix='_life_ex')
```

```python
# Merge the long tables for the chart frames
df_gapminder = df_fertility_long.merge(df_life_expectancy_long, on=['country', 'year'])
```

```python
df_gapminder = df_gapminder.merge(df_population_long, on=['country', 'year'])
```

```python
df_gapminder = df_gapminder.merge(df_continents.reset_index(), on=['country'], how='left')
```

```python
df_gapminder
```

```python
# Create chart frames
for year in range(1960, 2016):
    scatter_plot = sns.scatterplot(data=df_gapminder.loc[df_gapminder['year'] == year],
                    x='life expectancy',
                    y='fertility rate',
                    hue='continent',
                    size='population',
                    alpha=0.6)
    plt.axis((0, 85, 0, 10))
    plt.legend(loc=2, bbox_to_anchor=(1, 1));
    plt.title(year)
    figure = scatter_plot.get_figure()
    figure.savefig(f'./images/{year}.png',
                    bbox_inches='tight',
                    facecolor='white',
                    transparent=False)
    plt.close()
```
